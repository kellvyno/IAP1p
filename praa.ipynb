```python
#faça um programa que lê as duas notas obtidas por um aluno numa disciplina ao longo do semestre e calcule a sua media. a atribuiçao de conceitos obedece às regras:(MA= media de aproveitamento e  C = conceito) MA =  entre 9.0 e 10.0 C = A, MA = entre 7.5 e 8.99 C = B, MA entre 6.0 e 7.49 C = C, MD entre 4.0 e 5.99 C = D, entre 00.0 e 3.99 C= E. 
#o programa deve mostrar na tela as notas, a média, o conceito e a mensagem "aprovado" para conceitos A, B e C. "reprovado" se oconceito for D ou E.
# Programa para calcular média, conceito e aprovação de um aluno

# Solicitando as notas
nota1 = float(input("Digite a primeira nota (entre 0.0 e 10.0): "))
nota2 = float(input("Digite a segunda nota (entre 0.0 e 10.0): "))

media = (nota1 + nota2) / 2
if media >= 9.0:
  conceito = "A"
elif media >= 7.5:
  conceito = "B"
elif media >= 6.0:
  conceito = "C"
elif media >= 4.0:
  conceito = "D"
else:
  conceito = "E"

print(f"Notas: {nota1:.2f} e {nota2:.2f}")
print(f"Média: {media:.2f}")
print(f"Conceito: {conceito}")

if conceito in ["A", "B", "C"]:
  print("Aprovado!")
else:
  print("Reprovado.")

   
```

```python
#Faça um programa que leia  números e em seguida pergunte ao usúario qual a operação basica (soma, subtração, multiplicação, divisão,, exponenciação) ele deseja realizar. O resultado da operação deve ser acompanhado de uma frase que diga se o número é: par ou ímpar; positive ou negative; inteiro ou decimal.


numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))
operacao = input("Digite a operação desejada (+, -, *, /, ^): ")
if operacao == "+":
  resultado = numero1 + numero2
elif operacao == "-":
  resultado = numero1 - numero2
elif operacao == "*":
  resultado = numero1 * numero2
elif operacao == "/":
  if numero2 == 0:
    print("Erro: Divisão por zero.")
    exit()  # Encerra o programa
  else:
    resultado = numero1 / numero2
elif operacao == "^":
  resultado = numero1 ** numero2
else:
  print("Operação inválida.")
  
tipo_resultado = "decimal"
if resultado.is_integer():
  tipo_resultado = "inteiro"

paridade = "ímpar"
if resultado % 2 == 0:
  paridade = "par"

sinal = "positivo"
if resultado < 0:
  sinal = "negativo"
print(f"O resultado da operação {operacao} é {resultado:.2f}.")
print(f"O número é {tipo_resultado}, {sinal} e {paridade}.")
```

```python
#utilizando estruturas de repetição e listas(com ou sem .append), faça um programa que leia 5 números e informe a soma e a media dos números


numeros = []
for i in range(5):
  numero = float(input(f"Digite o {i+1}º número: "))
  numeros.append(numero)
for numero in numeros:
  soma += numero
media = soma / len(numeros)

print(f"A soma dos números é {soma:.2f}.")
print(f"A média dos números é {media:.2f}.")
```

```python
#faça um programa que peça umm número inteiro e deternine se ele é ou não número primo. Um número primo é aquele que é divisível somente por ele mesmo e por 1


numero = int(input("Digite um número inteiro: "))
if numero <= 1:
  print(f"{numero} não é um número primo.")
else:
  primo = True
  for i in range(2, numero):
    if numero % i == 0:
      primo = False
  if primo:
    print(f"{numero} é um número primo.")
  else:
    print(f"{numero} não é um número primo.")
```

```python
#o departamento estadual de meteorologia lhe contratou para desenvolver um programa que leia 5 temperaturas e informe ao final a menor e a maior temperaturas informadas, bem como a media das temperaturas

temperaturas = []
for i in range(5):
  temperatura = float(input(f"Digite a {i+1}ª temperatura (°C): "))
  temperaturas.append(temperatura)
menor_temperatura = min(temperaturas)
maior_temperatura = max(temperaturas)

#média
soma = 0
for temperatura in temperaturas:
  soma += temperatura
media = soma / len(temperaturas)
print(f"A menor temperatura registrada foi de {menor_temperatura:.1f}°C.")
print(f"A maior temperatura registrada foi de {maior_temperatura:.1f}°C.")
print(f"A média das temperaturas é de {media:.1f}°C.")
```

```python
""" 14. Faça um programa para o cálculo de uma folha de pagamento, sabendo que os descontos são do imposto de Renda, que depende do salário bruto (conforme tabela abaixo), 3% para o Sindicato e que o FGTS corresponde a 11% do salário bruto, mas não é descontado (é a empresa que deposita). O salário líquido corresponde ao salário bruto menos os descontos. O programa deverá pedir ao usuário o valor da sua hora e a quantidade de horas trabalhadas no mês.
Desconto do IR:
Salário Bruto até R$ 2.112,00(inclusive) - Isento;
De 2,112,01 até 2.826,65(inclusive) - desconto de 7.5%;
De 2.826,66 até 3.751,05(inclusive) - desconto de 15%;
De 3.751,06 até 4.664,68(inclusive) - desconto de 22.5%; Acima de 4.664,68 desconto de 27.5%."""

valor_hora = float(input("Digite o valor da sua hora: R$ "))
horas_trabalhadas = int(input("Digite a quantidade de horas trabalhadas no mês: "))
salario_bruto = valor_hora * horas_trabalhadas

if salario_bruto <= 2112.00:
  desconto_ir = 0
elif salario_bruto <= 2826.65:
  desconto_ir = salario_bruto * 0.075
elif salario_bruto <= 3751.05:
  desconto_ir = salario_bruto * 0.15
elif salario_bruto <= 4664.68:
  desconto_ir = salario_bruto * 0.225
else:
  desconto_ir = salario_bruto * 0.275

desconto_sindicato = salario_bruto * 0.03
fgts = salario_bruto * 0.11
salario_liquido = salario_bruto - desconto_ir - desconto_sindicato
print(f"Salário Bruto: R$ {salario_bruto:.2f}")
print(f"Desconto IR: R$ {desconto_ir:.2f}")
print(f"Desconto Sindicato: R$ {desconto_sindicato:.2f}")
print(f"FGTS (empresa): R$ {fgts:.2f}")
print(f"Salário Líquido: R$ {salario_liquido:.2f}")
```
